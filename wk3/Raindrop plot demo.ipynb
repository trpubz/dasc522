## Load dataset, DP

```
df = pd.read_excel('drug_consumption.xlsx',sheet_name='drug_consumption')
df = df.set_index('ID')

df2 = df[['THC', 'Age', 'Gender','Education', 'Country','Ethnicity','Nscore', 'Escore', 'Oscore', 'Ascore',\
          'Cscore', 'Impulsive', 'SS']]

df2.info()
df2.describe()
df2.head()
```

Raindrop plot

```
sns.set(style="whitegrid",font_scale=1.25)
f, ax = plt.subplots(figsize=(8,16))
sigma = .25
ax=pt.RainCloud(data = df2, palette = 'viridis',scale = "count",
      bw = sigma,width_viol = 0.6, orient = "h", move=0.2)
plt.show()
```
